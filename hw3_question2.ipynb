## (c) 

```python
# Gaussian blur
#img2 = cv2.GaussianBlur(img2, (5, 5), 0)
#plt.figure(figsize = (15,5))
#plt.imshow(img2)
```

I applied gaussian blur to make picture more smooth, so that I could reduce image noise, eliminate sharp edge and enhance image structure for better approximation

```python
# randomly sample 5000 pixels

train_data = []
train_label = []
for i in range(5000):
    x_r = random.randint(0,len(img2)-1)
    y_r = random.randint(0,len(img2[0])-1)

    train_data.append([x_r,y_r])
    train_label.append(img2[x_r,y_r]/255)

train_data = np.array(train_data)
train_label = np.array(train_label)
```

```python
# store coordinates of test data
test_data = []
test_label = []
for i in range(len(img2)):
    for j in range(len(img2[0])):
        test_data.append([i,j])
        test_label.append(img2[i,j]/255)

test_data = np.array(test_data)
test_label= np.array(test_label)
#test_data = np.reshape(img2, (543600,3))
```

```python
def convert_grayscale(y):
    for i in range(len(y)):

        gray_x = (y[i][0] * 0.299) + (y[i][1] * 0.587) + (y[i][2] * 0.114)

        y[i][0] = gray_x
        y[i][1] = gray_x
        y[i][2] = gray_x

    return y
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
clf = RandomForestRegressor(n_estimators = 10)
clf.fit(train_data,train_label[:,0])
prediction_r = clf.predict(test_data)

clf.fit(train_data,train_label[:,1])
prediction_g = clf.predict(test_data)


clf.fit(train_data,train_label[:,2])
prediction_b = clf.predict(test_data)
```
